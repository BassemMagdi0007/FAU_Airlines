```python
# Import libraries
import pandas as pd
from pulp import *
```

__*Low Season*__

```python
# Load the dataset and import it into dataframe.
# Determine number of work weeks 
week = ["Mon", "Tue", "Wen", "Thu", "Fri", "Sat", "Sun"]

file_path = '../../Files/fau_airlines_shifts.csv'
df_low_season = pd.read_csv(file_path, nrows=18)
```

```python
# Replace all X values with 1 and NAN with 0
df_low_season = df_low_season.fillna(0).applymap(lambda x: 1 if x == "X" else x)
```

```python
# DataSet for low season
df_low_season
```

```python
# Fetch the average passengers columns and put it in an array 
avg_passenger_no_columns = []
for day_of_week in range(1, 8):
    avg_passenger_no_columns.append(f'Avg_Passenger_No_{day_of_week}')
    
avg_passenger_no_columns
```

```python
# Fetch the Shifts column
shifts = df_low_season.drop(columns=avg_passenger_no_columns + ["Time Windows"]).values
shifts
```

```python
# Read the rest of the data from the the csv file

# Number of shifts
shift_num = shifts.shape[1]
print("shift_num:", shift_num)

# Number of time windows
time_windows = shifts.shape[0]
print("time_windows:", time_windows)

# Number of customers measured per time window
avg_passenger_no_low = df_low_season[avg_passenger_no_columns].values
print("avg_passenger_no_low:\n", avg_passenger_no_low)

# Service level during the low Season
avg_processing_time = 2.4  # in minutes
service_level_low = (1 / int(60 / 2.4))
#service_level_low = int(60 / 2.4)
print("service_level_low:", service_level_low)
```

```python
# Determine the decision variable low Season
# final goal is to find the optimal number of workers for each time slot of each day
num_workers_indexes = []
for day_of_week in range(0,7):
  for shift_index in range(shift_num):
    num_workers_indexes.append(f'{day_of_week}_{shift_index}')
    
num_workers = LpVariable.dicts("num_workers", num_workers_indexes, cat="Integer")
print(num_workers)
```

```python
# Create LP problem for low Season
prob1 = LpProblem("scheduling_workers", LpMinimize)
```

```python
# Add constraints to the problem for low Season
avg_passenger_no_low
print(avg_passenger_no_low)
for day_of_week in range(7):
  for t in range(time_windows):
    print("avg_passenger_no:", avg_passenger_no_low[t][day_of_week])
    prob1 += lpSum([shifts[t, j] * num_workers[f'{day_of_week}_{j}'] for j in range(shift_num)]) >= avg_passenger_no_low[t][day_of_week] * service_level_low
```

```python
# Solve the problem for low season
prob1.solve()
print("Status:", LpStatus[prob1.status])
```

```python
# Print the results for low season
for index in num_workers:
  index_parts = index.split('_')
  day_of_week = int(index_parts[0])
  day = week[day_of_week]
  shift = index_parts[1]
  print(f"The number of workers needed for day {day} shift {shift} is {int(num_workers[index].value())} workers")
```

__*High Season*__

```python
# Create a copy of the dataset
df_high_season = df_low_season.copy()

# During the high season, especially throughout the winter and summer breaks, the average number of passengers doubles
df_high_season[avg_passenger_no_columns] *= 2

# Set Wednesdays' values equal to Mondays' values
# "Wednesdays are as busy as Mondays"
df_high_season['Avg_Passenger_No_3'] = df_high_season['Avg_Passenger_No_1']

# Set Saturdays' values equal to Sundays' values
# "Sundays are as busy as Saturdays"
df_high_season['Avg_Passenger_No_7'] = df_high_season['Avg_Passenger_No_6']
```

```python
# DataSet for high season
df_high_season
```

```python
# Create LP problem for high Season

prob2 = LpProblem("scheduling_workers", LpMinimize)
```

```python
# Add constraints to the problem for high Season
avg_passenger_no_high
print(avg_passenger_no_high)
for day_of_week in range(7):
  for t in range(time_windows):
    print("avg_passenger_no:", avg_passenger_no_high[t][day_of_week])
    prob2 += lpSum([shifts[t, j] * num_workers[f'{day_of_week}_{j}'] for j in range(shift_num)]) >= avg_passenger_no_high[t][day_of_week] * service_level_high
```

```python
# Solve the problem for high season

prob2.solve()
print("Status:", LpStatus[prob2.status])
```
